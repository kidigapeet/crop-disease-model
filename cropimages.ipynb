```python
from icrawler.builtin import GoogleImageCrawler
import os

# Define a function to download images
def download_images(search_term, folder_name, num_images=50):
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)
    crawler = GoogleImageCrawler(storage={'root_dir': folder_name})
    crawler.crawl(keyword=search_term, max_num=num_images)

# Define crop types and disease/health status
categories = {
    "Healthy_Tomato": "healthy tomato plant",
    "Healthy_Maize": "healthy maize plant",
    "Healthy_Potato": "healthy potato plant",
    "Healthy_Wheat": "healthy wheat plant",
    "Healthy_Banana": "healthy banana plant",

    "Tomato_Blight": "tomato blight disease",
    "Maize_Leaf_Spot": "maize leaf spot disease",
    "Potato_Early_Blight": "potato early blight",
    "Wheat_Rust": "wheat rust disease",
    "Banana_Bacterial_Wilt": "banana bacterial wilt",
}

# Download images for each category
for folder, term in categories.items():
    print(f"Downloading: {term} into {folder}...")
    download_images(term, f"images/{folder}", num_images=50)

print("\nâœ… Download complete!")
```

```python
import os
import pandas as pd

# Define root folder of downloaded images
root_dir = "images"

# Prepare a list to store records
data = []

# Traverse through each subfolder
for folder in os.listdir(root_dir):
    folder_path = os.path.join(root_dir, folder)
    if os.path.isdir(folder_path):
        # Assign label
        label = "healthy" if folder.lower().startswith("healthy") else "unhealthy"

        # Collect image file paths
        for filename in os.listdir(folder_path):
            if filename.lower().endswith((".jpg", ".jpeg", ".png")):
                full_path = os.path.join(folder_path, filename)
                data.append({"image_path": full_path, "label": label})

# Create DataFrame and save to CSV
df = pd.DataFrame(data)
df.to_csv("plant_disease_dataset.csv", index=False)

print("âœ… CSV created: plant_disease_dataset.csv")
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras import layers, models
import os

# Set image size and batch size
IMG_SIZE = 224
BATCH_SIZE = 32 # Match batch size with the previous generator

# Data generators
# Assuming train_df and test_df DataFrames are already created in a previous cell
# using train_test_split from the original df.
augmentor = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

train_data = augmentor.flow_from_dataframe(
    dataframe=train_df,
    x_col='image_path',
    y_col='label',
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=True
)

val_data = ImageDataGenerator(rescale=1./255).flow_from_dataframe(
    dataframe=test_df, # Use test_df for validation data
    x_col='image_path',
    y_col='label',
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    shuffle=False # No need to shuffle validation data
)


# Load pre-trained base model
base_model = MobileNetV2(include_top=False, input_shape=(IMG_SIZE, IMG_SIZE, 3), weights='imagenet')
base_model.trainable = False  # freeze base layers

# Build the classifier
model = models.Sequential([
    base_model,
    layers.GlobalAveragePooling2D(),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(1, activation='sigmoid')  # binary output
])

# Compile
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train
history = model.fit(train_data, validation_data=val_data, epochs=5)
```

```python
# Evaluate the model
loss, accuracy = model.evaluate(val_data)

print(f"\nâœ… Test Loss: {loss:.4f}")
print(f"âœ… Test Accuracy: {accuracy:.4f}")
```

```python
# Plot training history
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Make predictions on the validation data
predictions = model.predict(val_data)

# Convert predictions to binary labels (0 or 1)
predicted_labels = (predictions > 0.5).astype(int)

# Get true labels from the validation data generator
true_labels = val_data.classes

# Get the filenames of the validation images
filenames = val_data.filenames

# Create a DataFrame to display predictions and true labels
results_df = pd.DataFrame({
    'filename': filenames,
    'true_label': true_labels,
    'predicted_label': predicted_labels.flatten()
})

# Map numerical labels back to original class names
label_map = {v: k for k, v in val_data.class_indices.items()}
results_df['true_label_name'] = results_df['true_label'].map(label_map)
results_df['predicted_label_name'] = results_df['predicted_label'].map(label_map)

# Display the first few results
print("âœ… Sample Predictions:")
display(results_df.head())
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Generate the confusion matrix
cm = confusion_matrix(true_labels, predicted_labels)

# Plot the confusion matrix
plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=label_map.values(), yticklabels=label_map.values())
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```
